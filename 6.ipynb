```python
##positional argument
##keyword argument
def hello(name, age=21):
    print("My name is {} and age is {}".format(name,age))
```

```python
hello("Meghali")
```

```python
hello()
```

```python
def hello(*args, **kwargs):
    print(args)
    print(kwargs)
```

```python
hello('Meghali', 'Khan', age=21, dob=2004)
```

```python
list=['Meghali', 'Khan']
dict_args={'age' : 21 , 'dob' : 2004}
```

```python
hello(list, dict_args)
```

```python
hello(*list, **dict_args)
```

```python
lst=[1,2,3,4,5,6,7,8]
```

```python
def evenoddsum(lst):
    even_sum=0
    odd_sum=0
    for i in lst:
        if i%2==0:
            even_sum=even_sum+i
        else:
            odd_sum=odd_sum+i
    return even_sum, odd_sum
```

```python
evenoddsum(lst)
```
