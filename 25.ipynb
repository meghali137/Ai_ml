DBSCAN

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1,2], [2,2], [2,3], [8,7], [8,8], [25,80]])
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
db = DBSCAN(eps=3, min_samples=2)

db.fit(X)
db.labels_
```

```python
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
## create a concentric circle dataset
X, _ = make_circles(n_samples=500, factor=.5, noise=0.05, random_state=4)

dbscan = DBSCAN(eps=0.12, min_samples=5)
clusters = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN clustering of concentric circles")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

k = 5  # same as min_samples
neighbors = NearestNeighbors(n_neighbors=k)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)

# take distance to kth neighbor
k_distances = np.sort(distances[:, k-1])

plt.plot(k_distances)
plt.xlabel("Points sorted by distance")
plt.ylabel(f"{k}-NN distance")
plt.title("k-distance plot")
plt.show()
```

OPTICS(Ordering points to identify the clustering structure)

```python
from sklearn.cluster import OPTICS

optics = OPTICS(min_samples=5)
labels = optics.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("OPTICS clustering")
plt.show()
```

```python
optics = OPTICS(min_samples=5, cluster_method='dbscan', eps=0.12)
labels = optics.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("OPTICS (DBSCAN-style clusters)")
plt.show()
```
