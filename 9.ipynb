OOPs in Python

```python
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
    def self_drive(self):
        return "This is a {} car".format(self.enginetype)
```

```python
car1 = Car(4,2,"petrol")
```

```python
car1.self_drive()
```

```python
car2 = Car(4,3,"diesel")
```

```python
print(car1.doors)
```

```python
print(car2.enginetype)
```

```python
dir(car1)
```

```python
car1.windows=5
car1.doors=2
```

```python
print(car1.windows)
```

```python
car2=Car()
```

```python
car2.enginetype="Petrol"
```

```python
print(car2.enginetype)
```

Python OOPs : Public, Protected and Private

```python
##All the class variables are public
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
```

```python
audi=Car(4,5,"Diesel")
```

```python
dir(audi)
```

```python
##protected variable
class Car:
    def __init__(self, windows, doors, enginetype):
        self._windows=windows
        self._doors=doors
        self._enginetype=enginetype
```

```python
audi = Car(2,3,"Diesel")
```

```python
dir(audi)
```

```python
audi._windows
```

```python
audi._windows=5
```

```python
audi._windows
```

```python
##protected variable can be accessed only in the sub class or child class
```

```python
class Truck(Car):
    def __init__(self, windows, doors, enginetype, horsepower):
        super().__init__(windows,doors,enginetype)
        self.horsepower=horsepower
```

```python
truck=Truck(4,4,"Diesel",4000)
dir(truck)
```

```python
truck._doors
```

```python
##private
class Car:
    def __init__(self, windows, doors, enginetype):
        self.__windows=windows
        self.__doors=doors
        self.__enginetype=enginetype
```

```python
audi=Car(4,5,"Diesel")
```

```python
dir(audi)
```

INHERITANCE

```python
##Car Blueprint

class Car():
    def __init__(self, windows, doors, enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
    def drive(self):
        print("The Person drives the Car")
```

```python
car=Car(4, 3, "Diesel")
```

```python
car.windows
```

```python
dir(car)
```

```python
car.drive()
```

```python
class audi(Car):
    def __init__(self, windows, doors, enginetype, enableai):
        super().__init__(windows, doors, enginetype)
        self.enableai=enableai
    def selfdrive(self):
        print("Audi supports self-driving")
```

```python
audiQ7 = audi(5,5,"diesel", True)
```

```python
dir(audiQ7)
```

```python
audiQ7.enginetype
```

```python
audiQ7.drive()
```

```python
audiQ7.selfdrive()
```

```python
audiQ7.enableai
```
