```python
## SEABORN

## Distribution Plots : 1. distplot  2. joinplot  3. pairplot  (practice problem on IRIS Dataset)

## Univariate analysis, Bivariate analysis
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")   ##.pivot(index="",columns="",values="")
```

```python
df.head()
```

```python
## Correlation with Heatmap
## A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. It is very important in Feature Selection
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))    ##, annot=True
```

```python
sns.heatmap(df.corr(numeric_only=True), cmap="crest")
```

```python
## JoinPlot

## A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plt, a 2D histogram or a 2D density plot.

## Univariate Analysis

##Univariate analysis means analyzing one variable at a time.
```

```python
sns.jointplot(x='tip', y='total_bill', data=df)
```

```python
df1=sns.load_dataset("penguins")
```

```python
sns.jointplot(data=df1, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind="hex")
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind="reg")
```

```python
sns.jointplot(data=df1, x="bill_length_mm", y="bill_depth_mm", height=5, ratio=2, marginal_ticks=True)
```

```python
sns.pairplot(df)
```

```python
## Displot

## Displot helps us to check the distribution of the columns feature
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

```python
## displot()  ##for visualizing univariate pr bivariate distribution of data
```

```python
sns.displot(data=df1, x="flipper_length_mm")
```

```python
sns.displot(data=df1, x="flipper_length_mm", kind="kde")
```

```python
sns.displot(data=df1, x="flipper_length_mm", kind="ecdf")
```

```python
sns.displot(data=df1, x="flipper_length_mm", kde=True)
```

```python
## bivariate

sns.displot(data=df1, x="flipper_length_mm", y="bill_length_mm")
```

```python
sns.displot(data=df1, x="flipper_length_mm", y="bill_length_mm", kind="kde")
```

```python
sns.displot(data=df1, x="flipper_length_mm", y="bill_length_mm", kind="kde", rug=True)
```

```python
sns.displot(data=df1, x="flipper_length_mm", hue="species", kind="kde")
```

```python
sns.displot(data=df1, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
sns.displot(data=df1, x="flipper_length_mm", hue="species", col="sex", kind="kde")
```

```python
sns.displot(
    data=df1, y="flipper_length_mm", hue="sex", col="species",
    kind="ecdf", height=4, aspect=.7,
)
```
