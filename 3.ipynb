```python
## Compute the x and y coordinates for the points on a sine curve

x=np.arange(0, 4*np.pi, 0.1)
y=np.sin(x)
plt.title('Sine wave form')
plt.plot(x,y,'g')
```

```python
#subplot()
# compute tge x and y coordinates for points on sine and cosine curves
x=np.arange(0, 5*np.pi, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'r')
plt.title('Cosine')
```

```python
## Bar PLot

x = [2,6,8,10]
y = [11,16,6,9]

x2 = [3,9,11,13]
y2 = [6,15,7,2]

plt.bar(x,y)
plt.bar(x2,y2, color='g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
## Histograms

a=np.array([23,43,12,56,87,34,98,76,3,44,23,59])
plt.hist(a, bins=15)
plt.title('Histogram')
```

```python
## Box Plot using Matplotlib

data = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plor
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
## Pie Chart

labels = 'Python', 'C++' , 'Java' , 'C'
sizes = [215, 130, 240, 205]
colors = ['gold', 'lightcoral', 'yellowgreen', 'lightskyblue']
explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
```
